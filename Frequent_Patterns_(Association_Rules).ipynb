```python
frequentItemsets = list(apriori(transactions,min_support = 0.065))
frequentItemsets
```

```python
min_sup = 0.03

frequentItemsets = list(apriori(transactions,min_support = min_sup))
for i in frequentItemsets:
    if i[0].__len__()>1:
        print('yes')
        print(i)
```
